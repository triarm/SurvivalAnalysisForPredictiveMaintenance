```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc = {'figure.figsize':(16,9)})
```

```python
column1 = ["machine_name", "cycle", "operational_setting_1", "operational_setting_2", "operational_setting_3"]
column2 = [f'sensor_measurement_{i:02}' for i in range(1,22)]
columns = column1 + column2
```

```python
turbofan_df = pd.read_csv("dataset/train_FD001.txt", header = None, sep = "\s+", names = columns)
turbofan_df.head()
```

```python
max_cycle = turbofan_df.groupby(by = "machine_name")['cycle'].transform(max)
turbofan_df = turbofan_df[turbofan_df["cycle"] == max_cycle]
turbofan_df.head()
```

```python
plt.hlines(y=turbofan_df['machine_name'], xmin=1, xmax=turbofan_df['cycle'], color='skyblue')
plt.plot(turbofan_df['cycle'], turbofan_df['machine_name'], "o")
plt.plot([1 for i in range(len(turbofan_df))], turbofan_df['machine_name'], "o")

# Add titles and axis names
plt.title("Max. Cycle")
plt.xlabel('Cycle')
plt.ylabel('Machine ID')

# Show the plot
plt.show()
```

```python
deleted_columns = list(turbofan_df.loc[:, (turbofan_df.nunique() == 1)].columns)
deleted_columns.append('sensor_measurement_17')
deleted_columns.append('operational_setting_2')
```

```python
deleted_columns = sorted(deleted_columns)
deleted_columns
```

```python
turbofan_df = turbofan_df[turbofan_df.columns[~turbofan_df.columns.isin(deleted_columns)]]
```

```python
# turbofan_df['cycle'] = turbofan_df['cycle'].apply(lambda x: x if x < 220 else 220)
turbofan_df['status'] = turbofan_df['cycle'].apply(lambda x: False if x > 220 else True)
```

```python
turbofan_df
```

```python
from sksurv.nonparametric import kaplan_meier_estimator

time, survival_prob = kaplan_meier_estimator(turbofan_df["status"], turbofan_df["cycle"])
time, survival_prob = np.append(0, time), np.append(1, survival_prob)
plt.step(time, survival_prob, where = "post")
plt.xlim(left = 0, right = 220)
plt.title("Kaplan Meier Estimator")
plt.ylabel("est. probability of survival $\hat{S}(t)$")
plt.xlabel("time $t$")
```

```python
data_x = turbofan_df.iloc[:, 2:16]
data_y = turbofan_df.iloc[:, [16,1]]
data_y = list(data_y.itertuples(index = None, name = None))
data_y = np.array(data_y, dtype=[('status', bool), ('cycle', int)])
```

```python
data_y
```

```python
sns.heatmap(data_x.corr(), annot = True,)
```

```python
from sksurv.linear_model import CoxPHSurvivalAnalysis

estimator = CoxPHSurvivalAnalysis(verbose = 10, alpha = 1e-5)
estimator.fit(data_x, data_y)
```
