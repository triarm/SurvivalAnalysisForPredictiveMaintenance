```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc = {'figure.figsize':(16,9)})
```

# Import Data and Preprocessing

```python
maintenance_df = pd.read_csv("dataset/maintenance.csv", sep = ";")
maintenance_df.head()
```

```python
maintenance_df[['team', 'provider']] = maintenance_df[['team', 'provider']].astype('category')
maintenance_df['broken'] = maintenance_df['broken'].astype('bool')
maintenance_df.info()
```

# Correlation Heatmap

```python
sns.heatmap(maintenance_df.corr(), annot = True,)
```

> Tidak ada korelasi yang tinggi antar variabel

# Data Preparation for Modeling

```python
# One Hot Encoding for Categorical Variable
from sksurv.preprocessing import OneHotEncoder

data_x = OneHotEncoder().fit_transform(maintenance_df.iloc[:, 2:])
data_x.head()
```

```python
# Preprocessing for target variable

data_y = list(maintenance_df.iloc[:, [1,0]].itertuples(index = None, name = None))
data_y = np.array(data_y, dtype=[('broken', bool), ('lifetime', float)])
data_y
```

# Kaplan Meier Estimator

```python
from sksurv.nonparametric import kaplan_meier_estimator

time, survival_prob = kaplan_meier_estimator(data_y["broken"], data_y["lifetime"])
time, survival_prob = np.append(0, time), np.append(1, survival_prob)
plt.step(time, survival_prob, where = "post")
plt.xlim(left = 0, right = 100)
plt.title("Kaplan Meier Estimator")
plt.ylabel("est. probability of survival $\hat{S}(t)$")
plt.xlabel("time $t$")
```

## Compare Survival Time for Each Team

```python
# Keplen Meier for every Team
for team in maintenance_df['team'].unique():
    team_df = maintenance_df[maintenance_df['team'] == team]
    time_treatment, survival_prob_treatment = kaplan_meier_estimator(team_df['broken'],
                                                                     team_df['lifetime'])
    plt.step(time_treatment, survival_prob_treatment, where = "post", label = team)
    
plt.ylabel("est. probability of survival $\hat{S}(t)$")
plt.xlabel("time $t$")
plt.legend(loc="best")
```

```python
from sksurv.compare import compare_survival

compare_survival(data_y, maintenance_df['team'], return_stats = True)
```

> Kesimpulan: *p-value* $ = 1.58e-11 < 0.05$ maka terdapat distribusi survival yang berbeda diantara ketiga team tersebut

## Compare Survival Time for each Provider

```python
# Keplen Meier for every Provider
for provider in maintenance_df['provider'].unique():
    provider_df = maintenance_df[maintenance_df['provider'] == provider]
    time_treatment, survival_prob_treatment = kaplan_meier_estimator(provider_df['broken'],
                                                                     provider_df['lifetime'])
    plt.step(time_treatment, survival_prob_treatment, where = "post", label = provider)
    
plt.ylabel("est. probability of survival $\hat{S}(t)$")
plt.xlabel("time $t$")
plt.legend(loc="best")
```

```python
compare_survival(data_y, maintenance_df['provider'], return_stats = True)
```

> Kesimpulan: *p-value* $ = 3.72e-170 < 0.05$ maka terdapat distribusi survival yang berbeda diantara mesin yang berasal dari keempat provider tersebut

# Cox Propotional Hazard

```python
from sksurv.linear_model import CoxPHSurvivalAnalysis

estimator = CoxPHSurvivalAnalysis()
estimator.fit(data_x, data_y)
```

```python
# Coef model
pd.Series(estimator.coef_, index = data_x.columns)
```

**Interpretation:**

1. Mesin yang dioperasikan TeamC meningkatkan tingkat kerusakan hingga 40x dibanding TeamA
2. Mesin yang berasal dari Provider3 memiliki tingkat kerusakan yang paling tinggi dibanding mesin dari provider lainnya

# Model Evaluation

```python
from sksurv.metrics import concordance_index_censored

prediction = estimator.predict(data_x)
result = concordance_index_censored(maintenance_df["broken"], maintenance_df["lifetime"], prediction)
result[0]
```

```python
estimator.score(data_x, data_y)
```
